# Dibujar un vector de forma alterna

```python
import numpy as np  # Importa la biblioteca numpy para realizar operaciones numéricas
import matplotlib.pyplot as plt  # Importa la biblioteca matplotlib para la creación de gráficos

# Dibuja un vector (flecha) con origen en (1, 2) y componentes (3, 4).
# scale_units='xy' indica que la escala se basa en las unidades del gráfico en los ejes x e y.
# scale=1 asegura que el vector no se ajuste en tamaño y se muestre en la escala de los ejes.
plt.quiver(1, 2, 3, 4, scale_units='xy', scale=1)

# Establece los límites del eje X de 0 a 4.
plt.xlim(0, 4)

# Establece los límites del eje Y de 0 a 8.
plt.ylim(0, 8)

# Muestra el gráfico (solo necesario si ejecutas el código en un entorno que no lo hace automáticamente)
plt.show()
```

## Dibujamos las Guias

```python
import matplotlib.pyplot as plt

# Dibuja un vector (flecha) desde (1, 2) con componentes (3, 4).
# scale_units='xy' y angles='xy' indican que tanto la escala como el ángulo
# se basan en las unidades del gráfico en los ejes X e Y.
# scale=1 asegura que el vector se muestre a tamaño real sin ajustes de escala.
plt.quiver(1, 2, 3, 4, scale_units='xy', angles='xy', scale=1)

# Establece los límites del eje X de 0 a 4.
plt.xlim(0, 4)

# Establece los límites del eje Y de 0 a 8.
plt.ylim(0, 8)

# Dibuja una línea vertical como guía en x=1 (origen del vector en X).
plt.axvline(x=1)

# Dibuja una línea horizontal como guía en y=2 (origen del vector en Y).
plt.axhline(y=2)

# Dibuja otra línea vertical en x=1+3 (posición final del vector en X).
plt.axvline(x=1+3)

# Dibuja otra línea horizontal en y=2+4 (posición final del vector en Y).
plt.axhline(y=2+4)

# Muestra el gráfico
plt.show()
```

## Graficar una lista

```python
import matplotlib.pyplot as plt

# Dibuja dos vectores:
# El primer vector empieza en (1, 2) con componentes (1, 3).
# El segundo vector empieza en (0, 0) con componentes (2, 2).
plt.quiver([1, 0], [2, 0], [1, 2], [3, 2], scale_units='xy', angles='xy', scale=1)

# Establece los límites del eje X de 0 a 4.
plt.xlim(0, 4)

# Establece los límites del eje Y de 0 a 6.
plt.ylim(0, 6)

# Dibuja líneas guía en los ejes X e Y para el primer vector.
# Línea vertical en x=1 y horizontal en y=2 (origen del primer vector).
plt.axvline(x=1, color='gray', linestyle='--', alpha=0.3)
plt.axhline(y=2, color='gray', linestyle='--', alpha=0.3)

# Dibuja líneas guía en los ejes X e Y para el segundo vector.
# Línea vertical en x=2 y horizontal en y=2 (final del segundo vector en el eje X).
plt.axvline(x=2, color='gray', linestyle='--', alpha=0.3)
plt.axhline(y=2, color='gray', linestyle='--', alpha=0.3)

# Otra línea guía para la posición final del primer vector.
# Línea horizontal en y=5 (final del primer vector en el eje Y).
plt.axhline(y=5, color='gray', linestyle='--', alpha=0.3)

# Muestra el gráfico
plt.show()
```

## Graficar Arreglos de Numpy

```python
import numpy as np  # Importa numpy para manejar vectores
import matplotlib.pyplot as plt  # Importa matplotlib para generar gráficos

# Definición de dos vectores
v1 = np.array([1, 2])  # Primer vector con componentes (1, 2)
v2 = np.array([3, 4])  # Segundo vector con componentes (3, 4)

# Dibuja dos vectores que parten desde el origen (0, 0)
# El primer vector es v1 con componentes (1, 2)
# El segundo vector es v2 con componentes (3, 4)
plt.quiver([0, 0], [0, 0], [v1[0], v2[0]], [v1[1], v2[1]], scale_units='xy', angles='xy', scale=1)

# Ajusta los límites del eje X e Y basándose en las componentes de los vectores
plt.xlim(0, max(v1[0], v2[0]))  # El límite del eje X es el máximo de los componentes X de v1 y v2
plt.ylim(0, max(v1[1], v2[1]))  # El límite del eje Y es el máximo de los componentes Y de v1 y v2

# Dibuja líneas guía para el primer vector (v1)
plt.axvline(x=1, color='gray', linestyle='--', alpha=0.3)  # Línea vertical en la componente X de v1
plt.axhline(y=2, color='gray', linestyle='--', alpha=0.3)  # Línea horizontal en la componente Y de v1

# Dibuja líneas guía para el segundo vector (v2)
plt.axvline(x=3, color='gray', linestyle='--', alpha=0.3)  # Línea vertical en la componente X de v2
plt.axhline(y=4, color='gray', linestyle='--', alpha=0.3)  # Línea horizontal en la componente Y de v2

# Muestra el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt  # Importa la biblioteca matplotlib para gráficos

# Definición de la función para graficar vectores
def graficar_vectores(vecs, cols):
    plt.figure(figsize=(5,5))  # Crea una figura de tamaño 5x5
    # 'vecs' es una lista de vectores, 'cols' es una lista de colores correspondientes
    for i in range(len(vecs)):  # Itera sobre la lista de vectores
        vec = vecs[i]  # Obtiene el vector en la posición i
        # Dibuja el vector que parte del origen (0, 0) con componentes vec[0] y vec[1]
        # El color del vector es el que corresponde en la lista 'cols' en la posición i
        plt.quiver(0, 0, vec[0], vec[1], color=cols[i], scale_units='xy', angles='xy', scale=1)

    plt.xlim(-10, 10)  # Ajusta el límite del eje X para una mejor visualización
    plt.ylim(-10, 10)  # Ajusta el límite del eje Y para una mejor visualización
    plt.grid(True)  # Añade una cuadrícula al gráfico para mayor claridad
    plt.show()  # Muestra el gráfico
```

**Uso de la funcion para Graficar**

```python
# Definición de dos vectores
v1 = np.array([1, 2])
v2 = np.array([3, 4])

# Llama a la función para graficar los vectores v1 y v2
graficar_vectores([v1, v2], ['blue', 'red'])

# Ajusta los límites de los ejes con base en los máximos de los vectores
plt.xlim(0, max(v1[0], v2[0]) + 1)  # Se añade un pequeño margen
plt.ylim(0, max(v1[1], v2[1]) + 1)  # Se añade un pequeño margen
plt.show()
```

```python
# Definición de tres vectores
v1 = np.array([1, 2])
v2 = np.array([3, 4])
v3 = np.array([8, 4])

# Llama a la función para graficar los vectores v1, v2 y v3
graficar_vectores([v1, v2, v3], ['blue', 'red', 'black'])

# Ajusta los límites de los ejes con base en los máximos de los vectores
plt.xlim(0, max(v1[0], v2[0], v3[0]) + 1)  # Añade un pequeño margen
plt.ylim(0, max(v1[1], v2[1], v3[1]) + 1)  # Añade un pequeño margen
plt.show()
```
