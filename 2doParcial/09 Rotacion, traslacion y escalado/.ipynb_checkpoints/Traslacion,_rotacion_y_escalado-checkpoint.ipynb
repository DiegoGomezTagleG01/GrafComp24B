## Rotaci√≥n

La rotaci√≥n es una transformaci√≥n que gira una figura geom√©trica alrededor de un punto fijo (t√≠picamente el origen) por un √°ngulo determinado. La forma y el tama√±o de la figura no cambian, solo su orientaci√≥n.

F√≥rmulas para la rotaci√≥n en un √°ngulo
ùúÉ
Œ∏ alrededor del origen:

**x‚Ä≤=x‚ãÖcos(Œ∏)‚àíy‚ãÖsin(Œ∏)**

**y‚Ä≤=x‚ãÖsin(Œ∏)+y‚ãÖcos(Œ∏)**

Donde Œ∏ es el √°ngulo de rotaci√≥n en radianes. Un √°ngulo positivo indica una rotaci√≥n en sentido antihorario y un √°ngulo negativo indica una rotaci√≥n en sentido horario.

```python
import numpy as np
import matplotlib.pyplot as plt

# Coordenadas del tri√°ngulo
x = [1, 2, 1.5, 1]
y = [1, 1, 2, 1]

# Grado de rotaci√≥n
theta = 198

# Funci√≥n para rotar la figura
def rotar(x, y, theta):
    # Convertir listas a arrays de numpy
    x = np.array(x)
    y = np.array(y)

    # Convertir grados a radianes
    theta_rad = np.radians(theta)

    # Aplicar las f√≥rmulas de rotaci√≥n
    xr = x * np.cos(theta_rad) - y * np.sin(theta_rad)
    yr = x * np.sin(theta_rad) + y * np.cos(theta_rad)

    return xr, yr

# Rotar el tri√°ngulo
xx, yy = rotar(x, y, theta)

# Graficar el tri√°ngulo original y rotado
plt.plot(x, y, label="Original")
plt.plot(xx, yy, label="Rotado")

plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
# Coordenadas del cuadrado
x = [1, 2, 2, 1, 1]  # Agregu√© el primer punto al final para cerrar la figura
y = [1, 1, 2, 2, 1]

# Grado de rotaci√≥n
theta = 198

# Funci√≥n para rotar la figura
def rotar(x, y, theta):
    # Convertir listas a arrays de numpy
    x = np.array(x)
    y = np.array(y)

    # Convertir grados a radianes
    theta_rad = np.radians(theta)

    # Aplicar las f√≥rmulas de rotaci√≥n
    xr = x * np.cos(theta_rad) - y * np.sin(theta_rad)
    yr = x * np.sin(theta_rad) + y * np.cos(theta_rad)

    return xr, yr

# Rotar el cuadrado
xx, yy = rotar(x, y, theta)

# Graficar el cuadrado original y rotado
plt.plot(x, y, label="Original")
plt.plot(xx, yy, label="Rotado")

plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

## Escalamiento

El escalamiento es una transformaci√≥n que cambia el tama√±o de una figura geom√©trica sin alterar su forma. Esta transformaci√≥n puede ser uniforme (cuando las dimensiones se modifican en la misma proporci√≥n en todas las direcciones) o no uniforme (cuando se cambia el tama√±o de manera diferente en cada direcci√≥n)

F√≥rmulas para el escalamiento:

ùë•‚Ä≤=ùë†ùë•‚ãÖùë•

ùë¶‚Ä≤=ùë†ùë¶‚ãÖùë¶y

Donde
ùë†ùë•
  y
ùë†ùë¶
  son los factores de escala en los ejes ùë• ùë¶

```python
#coordenadas de la figura origial

c = [1, 2, 1.5, 1]
r = [1, 1, 2, 1]

#el escalamiento
Sx, Sy = 4, 4

#formula

cc =  [Sx*x for x in c]
rr = [Sy*y for y in r]

#muestra original y escalonada

plt.plot(c,r)
plt.plot(cc,rr)
```
