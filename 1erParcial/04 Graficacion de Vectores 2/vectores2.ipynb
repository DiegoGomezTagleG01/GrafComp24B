### vector B

```python
import numpy as np
import matplotlib.pyplot as plt

# Límites del gráfico
x1 = -10
x2 = 10
y1 = -10
y2 = 10

plt.axis([x1, x2, y1, y2])
plt.axis('on')
plt.grid(True)
plt.title("Vector (-5, 2)")

# Graficar puntos de referencia
dx = 5
dy = 5
for x in np.arange(x1, x2, dx):
    for y in np.arange(y1, y2, dy):
        plt.scatter(x, y, s=1.5, color='lightgray')

# Graficar el vector (-5, 2) desde el origen (0,0)
plt.arrow(0, 0, -5, 2, head_length=0.5, head_width=0.3, color="k")

# Graficar las líneas punteadas de las componentes
# Línea punteada horizontal (componente x)
plt.plot([-5, -5], [0, 2], linestyle='--', color='black')

# Línea punteada vertical (componente y)
plt.plot([0, -5], [0,0], linestyle='--', color='black')

# Agregar texto en cursiva debajo de las líneas punteadas
plt.text(-5.3, -0.5, 'componente en X', fontsize=10, color='blue', style='italic')
plt.text(-0.5, 2.3, 'Componente en Y', fontsize=10, color='blue', style='italic')

# Agregar texto en cursiva encima del vector
plt.text(-6, 1.5, 'Vector(-5, 2)}', fontsize=12, color='red')

plt.show()
```
