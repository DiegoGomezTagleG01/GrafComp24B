## Primera parte 

Grafica los vetores dados

A=<2,4>
B=<5,0>
C=<0,-4>
D=<-3,-5>

```python
import numpy as np
import matplotlib.pyplot as plt

# Límites del gráfico
x1 = -10
x2 = 10
y1 = -10
y2 = 10

plt.axis([x1, x2, y1, y2])
plt.axis('on')
plt.grid(True)
plt.title("Vectores")

# Graficar puntos de referencia
dx = 5
dy = 5
for x in np.arange(x1, x2, dx):
    for y in np.arange(y1, y2, dy):
        plt.scatter(x, y, s=1.5, color='lightgray')

plt.arrow(0, 0, 2, 4, head_length=0.5, head_width=0.3, color="k")
plt.arrow(0, 0, 5, 0, head_length=0.5, head_width=0.3, color="y")
plt.arrow(0, 0, 0, -4, head_length=0.5, head_width=0.3, color="r")
plt.arrow(0, 0, -3, -5, head_length=0.5, head_width=0.3, color="b")



plt.show()
```

## Combinaciones M

```python
import numpy as np
import matplotlib.pyplot as plt

# Límites del gráfico
x1 = -10
x2 = 10
y1 = -10
y2 = 10

plt.axis([x1, x2, y1, y2])
plt.axis('on')
plt.grid(True)
plt.title("Vectores M")

# Graficar puntos de referencia
dx = 5
dy = 5
for x in np.arange(x1, x2, dx):
    for y in np.arange(y1, y2, dy):
        plt.scatter(x, y, s=1.5, color='lightgray')

plt.arrow(2, 4, -3, 3, head_length=0.5, head_width=0.3, color="k") #MN
plt.arrow(2, 4, 1, -4, head_length=0.5, head_width=0.3, color="k") #MP


plt.show()
```
