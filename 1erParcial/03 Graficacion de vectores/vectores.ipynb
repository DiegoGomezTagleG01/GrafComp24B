## Dibuja los puntos en el plano

```python
import numpy as np
import matplotlib.pyplot as plt
x1=0
x2=80

y1=0
y2=80

plt.axis([x1,x2,y1,y2])
plt.axis('on')
plt.grid(True)
plt.title("Ejes de las coordenadas")

dx=8
dy=8
#Graficar puntos
for x in np.arange(x1,x2, dx):
    for y in np.arange(y1,y2,dy):
        plt.scatter(x,y, s=1.5, color= 'lightgray')
plt.show()
```

## Vectores Realizados en clase

```python
import numpy as np
import matplotlib.pyplot as plt
x1=0
x2=10

y1=0
y2=10

plt.axis([x1,x2,y1,y2])
plt.axis('on')
plt.grid(True)
plt.title("Vector PN")

dx=5
dy=5
#Graficar puntos
for x in np.arange(x1,x2, dx):
    for y in np.arange(y1,y2,dy):
        plt.scatter(x,y, s=1.5, color= 'lightgray')
# Graficar el vector
        #x , y, incremento, abscisa, longitud, ancho, color
plt.arrow(2, 4,3,-1, head_length=2, head_width=0.5, color="k")
plt.show()
```
