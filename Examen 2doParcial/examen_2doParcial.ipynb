```python
import numpy as np
import matplotlib.pyplot as plt

F = np.array([
    (0, 0, 1),  
    (1, 0, 1),
    (1, 2, 1),
    (2, 2, 1),
    (2, 3, 1),
    (1, 3, 1),
    (1, 4, 1),
    (3, 4, 1),
    (3, 5, 1),
    (0, 5, 1),
    (0, 0, 1)
])
theta = 5* np.pi/4 
rotacion = np.array([[np.cos(theta), np.sin(theta), 0.],
              [-np.sin(theta), np.cos(theta), 0.],
              [0., 0., 1.]])
reflexionX = np.array([[1., 0, 0], [0, -1., 0], [0., 0., 1.]])
tx, ty = 3, 2 
T = np.array([[1., 0, tx], [0, 1., ty], [0., 0., 1.]])

transformaciones = {
    "Rotación": rotacion,
    "Reflexión X": reflexionX,
    "Traslación": T
}

fig, axes = plt.subplots(1, 3, figsize=(15, 10))
axes = axes.flatten()

for idx, (nombre, matrix) in enumerate(transformaciones.items()):
    f_transformada = np.dot(F, matrix.T)
    print(nombre)
    print(f_transformada)
    
    
    ax = axes[idx]
    ax.plot(F[:, 0], F[:, 1], color="blue", label="Original")
    ax.plot(f_transformada[:, 0], f_transformada[:, 1], color="red", label="Transformado")
    ax.set_title(nombre)
    ax.set_aspect('equal', adjustable='box')
    ax.grid()
    ax.legend()
plt.tight_layout()
plt.show()
```
