## Suma de vectores

La suma de vectores se realiza componente a componente, sin embargo, se puede realizar de forma grafica dibujando el siguiente vector donde termina el otro

```python
import numpy as np
import matplotlib.pyplot as plt

# Límites del gráfico
x1 = -10
x2 = 10
y1 = -10
y2 = 10

plt.axis([x1, x2, y1, y2])
plt.axis('on')
plt.grid(True)
plt.title("Vector A+B+C")

# Graficar puntos de referencia
dx = 5
dy = 5
for x in np.arange(x1, x2, dx):
    for y in np.arange(y1, y2, dy):
        plt.scatter(x, y, s=1.5, color='lightgray')

# Graficar el vector A desde el origen (0,0)
plt.arrow(0, 0, 2, 4, head_length=0.5, head_width=0.3, color="k")
# Graficar el vector B
plt.arrow(2, 4, -6, -2, head_length=0.5, head_width=0.3, color="b")
# Graficar el vector C 
plt.arrow(-4, 2, 2, -7, head_length=0.5, head_width=0.3, color="r")
# Graficar el vector A+B+C
plt.arrow(0, 0, -2, -5, head_length=0.5, head_width=0.3, color="g")

# Agregar texto en cursiva encima del vector
plt.text(-6, -1, 'C <2,-7>', fontsize=12, color='red')

plt.text(3, 2, 'A <2,4>', fontsize=12, color='black')


plt.text(-2, 4, 'B <-6,-2>', fontsize=12, color='blue')


plt.text(0, -3, 'A+B+C <-2,-5>', fontsize=12, color='green')

plt.show()
```
